```python
import numpy as np
import matplotlib.pyplot as plt

# Define time range
t = np.linspace(0, 20, 400)

# Define signal with short-term and long-term components
u = np.sin(t) + np.cos(t / 10)

# Define basis functions for decomposition (simplified example)
K0 = np.ones_like(t)
K1 = np.exp(-t)
K2 = 0.5 * (3 * np.exp(-2*t) - 1)

# Approximate coefficients for decomposition (for illustration)
coeff0 = np.trapz(u * K0 * np.exp(-t), t) / np.trapz(K0**2 * np.exp(-t), t)
coeff1 = np.trapz(u * K1 * np.exp(-t), t) / np.trapz(K1**2 * np.exp(-t), t)
coeff2 = np.trapz(u * K2 * np.exp(-t), t) / np.trapz(K2**2 * np.exp(-t), t)

# Reconstruct the signal using basis functions
u_approx = coeff0 * K0 + coeff1 * K1 + coeff2 * K2

# Plot the original signal and its approximation
plt.figure(figsize=(12, 6))
plt.plot(t, u, label='Original Signal')
plt.plot(t, u_approx, label='Approximated Signal', linestyle='--')
plt.xlabel('Time t')
plt.ylabel('Signal value')
plt.title('Original Signal and Its Approximation Using Basis Functions')
plt.legend()
plt.grid(True)
plt.show()
```
